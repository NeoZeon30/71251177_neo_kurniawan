```python
#algoritma untuk menentukan jenis akar dari suatu persamaan kuadrat.
#Deklarasi
var int a,b,c

#Deskripsi
input a,b,c

d <- b ** 2 - 4 * a * c

if d >= 0 THEN
  print "Akar Real atau Nyata"
if d > 0 THEN
  print "Akar Real dan Berlainan"
else d = 0 THEN
  print "Akar Real dan Kembar"
else d < 0 THEN
  print "Akar Imajiner atau Tidak Real atau Khayal."
END IF
```
